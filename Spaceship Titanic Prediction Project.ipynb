# Introduction

In this [competition](https://www.kaggle.com/competitions/spaceship-titanic/data?select=train.csv) our goal is to predict whether a passenger was transported to an alternative dimension during the Spaceship Titanic's collision with the spacetime anomaly. To help make these predictions, we're given a set of personal records recovered from the ship's damaged computer system. 

These are the features we are going to work with:
- `PassengerId` - A unique Id for each passenger. Each Id takes the form gggg_pp where gggg indicates a group the passenger is travelling with and pp is their number within the group. People in a group are often family members, but not always.
- `HomePlanet` - The planet the passenger departed from, typically their planet of permanent residence. 
- `CryoSleep` - Indicates whether the passenger elected to be put into suspended animation for the duration of the voyage. Passengers in cryosleep are confined to their cabins.
- `Cabin` - The cabin number where the passenger is staying. Takes the form `deck/num/side`, where `side` can be either `P` for Port or `S` for Starboard.
- `Destination` - The planet the passenger will be debarking to.
- `Age` - The age of the passenger. 
- `VIP` - Whether the passenger has paid for special VIP service during the voyage. 
- `RoomService`, `FoodCourt`, `ShoppingMall`, `Spa`, `VRDeck` - Amount the passenger has billed at each of the Spaceship Titanic's many luxury amenities.
- `Name` - The first and last names of the passenger.
- `Transported` - Whether the passenger was transported to another dimension. This is the target, the column you are trying to predict.

A couple things to do:

```python
def clean_HomePlanet(df):
    nans = df['HomePlanet'].isna()
    length = sum(nans)
    replacement = random.choices(['Earth', 'Europa', 'Mars'], 
                                 weights=[0.541930, 0.247786, 0.210284], 
                                 k= length)
    df.loc[nans, 'HomePlanet'] = replacement
    
    return df
    
train = clean_HomePlanet(train)
```

```python
def clean_CryoSleep(df):
    nans = df['CryoSleep'].isna()
    length = sum(nans)
    replacement = random.choices([True, False], 
                                weights= [0.368306, 0.641694],
                                k= length)
    df.loc[nans, 'CryoSleep'] = replacement
    
    return df

train = clean_CryoSleep(train)
```

```python
def clean_Destination(df):
    nans = df['Destination'].isna()
    length = sum(nans)
    replacement = random.choices(['TRAPPIST-1e', '55 Cancri e', 'PSO J318.5-22'], 
                                 weights = [0.694983, 0.211491, 0.093526], 
                                 k= length)
    df.loc[nans, 'Destination'] = replacement
    
    return df

train = clean_Destination(train)
```

```python
def clean_Age(df):
    nans = df['Age'].isna()
    length = sum(nans)
    replacement = random.choices([19, 27, 38, 79],
                                weights = [0.25, 0.25, 0.25, 0.25],
                                k= length)
    df.loc[nans, 'Age'] = replacement
    
    return df

train = clean_Age(train)
```

```python
def clean_other_columns(df):
    
    # Replace the NaN values of several columns with the most frequent value
    df['Cabin'] = df['Cabin'].fillna('Unknown')
    df['RoomService'] = df['RoomService'].fillna(0)
    df['FoodCourt'] = df['FoodCourt'].fillna(0)
    df['ShoppingMall'] = df['ShoppingMall'].fillna(0)
    df['Spa'] = df['Spa'].fillna(0)
    df['VRDeck'] = df['VRDeck'].fillna(0)
    
    # Drop VIP and Name columns
    df = df.drop(['VIP', 'Name'], axis=1)
    
    return df

train = clean_other_columns(train)
```

## Group

The `PassengerId` is in the form gggg-pp where gggg is the group they belong to, which is usually their family. PP is the passenger's number within their group. Therefore, we're going to separate the group from the passengerid in this section. 

```python
def separate_group(df):
    df['Group'] = df['PassengerId'].apply(lambda x: x.split('_')[0]).astype(int)
    df['Passenger Number'] = df['PassengerId'].apply(lambda x: x.split('_')[1]).astype(int)
    
    # Drop PassengerId. This column is not useful as of right now. 
    df = df.drop(['PassengerId'], axis=1)
    
    return df

train = separate_group(train)
train.head()
```

```python
print("The number of unique groups: {}".format(train.Group.nunique()))
```

```python
print("The number of unique passenger numbers: {}".format(train['Passenger Number'].nunique()))
```

There are many unique values. This will get very messy if we make this into a dummy variable. 

I think it's best to create a variable called `Is_Alone` that will be 1 if the passenger is the only person in his/her group and 0 if there are more people in the same group. (Of course this is assuming that groups are passengers traveling together.)

```python
train.info()
```

```python
def is_alone(df):
    group_freq = {}

    for index, row in df.iterrows():
        group = row['Group']
    
        if group in group_freq:
            group_freq[group] += 1
        else:
            group_freq[group] = 1
            
    df['Is_Alone'] = df['Group'].apply(lambda g: 0 if group_freq[g] > 1 else 1)
    
    # No longer need the group so let's drop this
    df.drop('Group', axis=1, inplace=True)
    
    return df

train = is_alone(train)
train.head()
```

## Money Spent Onboard

```python
train.describe()[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].transpose()
```

Let's add the following 3 features:
- `Total Spent` - The sum spent across all amenities 
- `Amenities Used` - The number of amenities used.
- `Max Spent Amenity` - The amenity with the largest spent.
- `Min Spent Amenity` - The amentity with the lowest spent.


```python
def amenities_used(df):
    
    for col in ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']:
        
        df[col + '_Used'] = df[col].apply(lambda x: 1 if x>0 else 0)
        
    df['Amenities Used'] = df[['RoomService_Used', 'FoodCourt_Used', 
                        'ShoppingMall_Used', 'Spa_Used', 'VRDeck_Used']].sum(axis=1)
    
    df = df.drop(['RoomService_Used', 'FoodCourt_Used', 'ShoppingMall_Used', 'Spa_Used', 'VRDeck_Used'], axis=1)
    
    return df

train = amenities_used(train)
```

```python
def max_spent_amenity(df):
    df['Max Spent Amenity'] = df[['RoomService', 'FoodCourt', 
                            'ShoppingMall', 'Spa', 'VRDeck']].idxmax(axis=1)
    
    return df

train = max_spent_amenity(train)
```

```python
def min_spent_amenity(df):
    df['Min Spent Amenity'] = df[['RoomService', 'FoodCourt', 
                                 'ShoppingMall', 'Spa', 'VRDeck']].idxmin(axis=1)
    
    return df

train = min_spent_amenity(train)
```

```python
def total_spent(df):
    df['Total Spent'] = df[['RoomService', 'FoodCourt', 
                            'ShoppingMall', 'Spa', 'VRDeck']].sum(axis=1)
    
    return df

train = total_spent(train)
```

```python
train.head()
```

Looks like the max spent and min spent amentity have a value even when nothing was spent. Let's correct this by looking at the `total spent`, if it's 0 then we will replace the `max spent amenity` and `min spent amenity` with the string `N/A`

```python
def fixing_max(row):
    max_spent_amenity, total_spent = row[0], row[1]
    
    if total_spent == 0:
        return 'N/A'
    else:
        return max_spent_amenity
    
        
train['Max Spent Amenity'] = train[['Max Spent Amenity', 'Total Spent']].apply(fixing_max, axis=1)
```

```python
def fixing_min(row):
    min_spent_amenity, total_spent = row[0], row[1]
    
    if total_spent == 0:
        return 'N/A'
    else:
        return min_spent_amenity
    
train['Min Spent Amenity'] = train[['Min Spent Amenity', 'Total Spent']].apply(fixing_min, axis=1)
```

```python
def create_dummies(df, dummy_cols):
    dummy_df = pd.get_dummies(df[dummy_cols], drop_first= True)
    df = pd.concat([df, dummy_df], axis=1)
    df = df.drop(dummy_cols, axis=1)
    
    return df

dummy_cols = ['HomePlanet', 'CryoSleep', 'Destination', 'Deck', 'Side',
              'Age_Categories', 'Max Spent Amenity', 'Min Spent Amenity']

train = create_dummies(train, dummy_cols)
train.head()
```

# Rescale the Data

Most machine learning models work best when all values are within the same domain. That domain is usually (0,1). Therefore, let's rescale the following values:
- Total Spent
- Amenities Used

To rescale we can use a module from the sklearn library

```python
from sklearn.preprocessing import minmax_scale

def rescale(df, cols):
    for col in cols:
        df[col + '_scaled'] = minmax_scale(df[col])
        
    df = df.drop(cols, axis=1)
        
    return df

train = rescale(train, ['Total Spent', 'Amenities Used', 'RoomService',
                        'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Age', 'Passenger Number'])

train.head()
```

# Choosing Best Features

We will be using the feature_selection.RFECV class which performs recursive feature elimination with cross-validation.

The RFECV class starts by training a model using all of your features and scores it using cross validation. It then uses the logit coefficients to eliminate the least important feature, and trains and scores a new model. At the end, the class looks at all the scores, and selects the set of features which scored highest.

```python
from sklearn.feature_selection import RFECV
```

```python
def optimized_features(df, model):
    all_X = df.drop('Transported', axis=1)
    all_y = df['Transported']

    selector = RFECV(model, cv=10)
    selector.fit(all_X, all_y)
    best_features = all_X.columns[selector.support_]
    
    return best_features
```

# Choosing the Best Model

`Grid-search` is a method that trains a number of models across a 'grid' of values and then searched for the model that gives us the highest accuracy.

Scikit-learn has a class to perform grid search, `model_selection.GridSearchCV()`. The 'CV' in the name indicates that we're performing both grid search and cross validation at the same time. 

By creating a dictionary of parameters and possible values and passing it to the `GridSearchCV` object you can automate the process. Our final step is to print the `GridSearchCV.best_params_` and `GridSearchCV.best_score_` attributes to retrieve the parameters of the best performing model, and the score it achieved. 

We can also use `GridSearchCV` to try combinations of different hyperparameters. Say we wanted to test values of `ball_tree`, `kd_tree`, and `brute` for the algorithm parameter and values `1`, `3`, and `5` for the n_neighbors algorithm parameter. `GridSearchCV` would train and test 9 models (3 for the first hyperparameter times 3 for the second hyperparameter). 

Let's use `GridSearchCV` to turbo-charge our search for the best performing parameters for our model, by testing many combinations of different hyperparameters. 

## KNearestNeighbors

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyperparameters = {
    "n_neighbors" : range(1,28,2),
    "weights" : ["distance", "uniform"],
    "algorithm" : ["brute"],
    "p": [1,2]
}
```

```python
def grid_search_ml(df, cols, model, hyperparameters):

    all_X = df[cols]
    all_y = df['Transported']
    grid = GridSearchCV(model, param_grid=hyperparameters, cv=10)
    grid.fit(all_X, all_y)
    
    best_params = grid.best_params_
    best_score = grid.best_score_
    
    return best_params, best_score
```

```python
best_params, best_score = grid_search_ml(train, all_features, KNeighborsClassifier(), hyperparameters)
print(best_params)
print(best_score)
```

```python
df.loc[len(df)] = ['KNearestNeighbors with Best Features and Best Model', best_score]
df
```

# Principal Component Analysis

**What Is It?**
- PCA is a dimensionality reduction technique that transforms a number of possibly correlated variables into a smaller number of uncorrelated variables; called principal components.

- The goal of PCA is to extract the most important information from the data table by compressing the size of the data set and keeping only the important information; then PCA computes new variables called principal components.

**Additional Info**
- 95% explained variance should be the criterium when choosing the number of principal components.
